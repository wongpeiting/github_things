```python
# it's always latitude first, longitude second. 
# so g.latlng[0] is latitude, g.latlng[1] is longitude.

# now i got to somehow convert the latitudes and longitudes into a dictionary. >> lesson here: https://littlecolumns.com/learn/python/getting-fancy-with-dictionaries/

import geocoder

for entry in data:
    address = f'{entry['street']}, {entry['city']}, {entry['state']}, {entry['postal_code']}, {entry['country']}'
    g = geocoder.arcgis(address)
    entry['lat'] = g.latlng[0]
    entry['lng'] = g.latlng[1]
    entry['lon'] = entry.pop('lng')

print(data)
```
